```python
listings.host_since = pd.to_datetime(listings.host_since)
```

```python
paris_data  = listings[listings.city == 'Paris']
```

```python
paris_data.head()
```

```python
paris_data.columns
```

```python
paris_data = paris_data.loc[:,['host_since','neighbourhood','city', 'accommodates', 'price']]
```

```python
paris_data.info()
```

```python
paris_data[paris_data.host_since.isna()]
```

```python
#since missing data is less than  0.001 percent so we can drop with out losing any valuable insight.
```

```python
paris_data.dropna(inplace=True)
```

```python
paris_data.info(memory_usage='deep')
```

```python
paris_data.describe().round()
```

```python
paris_data = paris_data.astype({'accommodates':'int8', 'price':'int16'})
```

```python
paris_listing_neighbourhood =  paris_data.groupby('neighbourhood')['price'].mean().sort_values()
paris_listing_neighbourhood
```

```python
paris_listings_accomodations =paris_data[paris_data.neighbourhood =='Elysee'].groupby(['accommodates'])['price'].mean().sort_values()
paris_listings_accomodations
```

```python
paris_data.groupby([paris_data.host_since.dt.year])['price'].mean().sort_values(ascending = True)
```

```python
paris_data.groupby([paris_data.host_since.dt.year]).agg({'price':'count'}).sort_values('price',ascending = True)
```

```python
paris_listing_neighbourhood.plot.barh(x = 'price', y ='neighbourhood', color= 'purple')
plt.xlabel('Price')
plt.ylabel('Neighbourhood')
plt.title('Paris listing avg prices by neighbourhood')
plt.legend()
plt.savefig("Paris listing avg prices by neighbourhood.png") 
plt.show()
```

```python
paris_listings_accomodations.plot.barh(color ='Teal')
plt.xlabel('Price')
plt.ylabel('Accomodations')
plt.title('Average price in Elysee Neighbourhood')
plt.legend()
plt.savefig("Average price in Elysee Neighbourhood'.png") 
plt.show()
```

```python
avg_price_over_time = paris_data.groupby([paris_data.host_since.dt.year])['price'].mean().sort_index(ascending = True)
avg_price_over_time = pd.DataFrame(avg_price_over_time)
avg_price_over_time
```

```python
growth_over_time =(paris_data.groupby([paris_data.host_since.dt.year])['price'].count() /
 paris_data.groupby([paris_data.host_since.dt.year])['price'].count().diff())
growth_over_time.name = 'growth_over_time'
growth_over_time = pd.DataFrame(growth_over_time)
growth_over_time
```

```python
pd.concat([avg_price_over_time,growth_over_time], axis = 1).iloc[6:13].plot()
```

Notes: after 2015 regulations price has been increased and there is steep decline in hosts 
       and picked up again after decrease in price 2020

```python
pd.concat([avg_price_over_time,growth_over_time], axis = 1).plot(y ='price')
```

## After 2015 regulations price has been increased due to surge in No of hosts preferring Air BnB.

```python
pd.concat([avg_price_over_time,growth_over_time], axis = 1).plot(y ='growth_over_time')
```

## Prices been affected in 2020 due to pandameic people were quarantined.

```python
dual_axis = pd.concat([avg_price_over_time,growth_over_time], axis = 1).reset_index()
dual_axis.head()
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
fig, ax1 = plt.subplots()
x = dual_axis.host_since
y1 = dual_axis.price
y2= dual_axis.growth_over_time

ax1.plot(x, y1, 'b-')
ax2 = ax1.twinx()


ax1.set_xlabel('years')
ax1.set_ylabel('Price')
ax2.set_ylabel('New Hosts over Time', color='r')

ax2.plot(x, y2, 'r-')

plt.title('New hosts VS average price over time')
plt.savefig("New hosts and average price over time.png") 
plt.show()
```
